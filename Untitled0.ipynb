```python
# Get and display feature mutual information
X_raw = flight_raw.copy()
# print(X_raw)
y_raw = X_raw.pop('Delay')
_ = X_raw.pop('id')

y_raw = y_raw.replace({np.nan: 0})


for col in X_raw.select_dtypes('object'):
        X_raw[col], _ = X_raw[col].factorize()

X_raw = X_raw.replace({np.nan: 0})

discrete_list = X_raw.dtypes == int

# print(X_raw)

mi_raw = mutual_info_classif(X_raw, y_raw, discrete_features = discrete_list)

mi_scores = pd.Series(mi_raw, name= "Mutual Information Scores", index = X_raw.columns).sort_values(ascending=True)

print(mi_raw)
print(mi_scores)

#mi_width = np.arange(len(mi_scores))
#mi_ticks = list(mi_scores.index)

#plt.figure(dpi=100, figsize=(8,5))
#plt.barh(mi_width, mi_scores)
#plt.yticks(mi_width, mi_ticks)
#plt.title("Mutual Information Scores")
```
